## 1: K means clustering

Points are (0,0) (0, 5), (6, 7) and (7, 0)

### a.

Initialize means at (0,0) and (7,0)


What means does this converge to?: 

(0,2.5) [(0,0) (0, 5)] and 

(6.5,3.5) [6, 7),(7, 0)]


### b.

Initialize means at (3,3) and (7,0)

What means does this converge to?: 

(2,4) [(0,0) (0, 5),(6,7)] and 

(7,0) [(7, 0)]


```python
def meanSqDistance(points, cassignments, cpoints):
    total=0
    n=points.shape[0]
    for i in range(0,n):
        point=points[i]
        cluster=cpoints[cassignments[i]]        
        dif=cluster-point
        error=np.sqrt(np.dot(dif, dif.T))
        total=total+error
    return total


points=np.array([[0,0],[0,5],[6,7],[7,0]])
cassA=[0,0,1,1]
cassB=[0,0,0,1]
cpointsA=[[0,2.5],[6.5,3.5]]
cpointsB=[[2,4],[7,0]]
```

```python
print(meanSqDistance(points,cassA,cpointsA))
print(meanSqDistance(points,cassB,cpointsB))
```

## Part b seeding

My proposed method of intiailizing seeds is to first pick a random data point as a seed, then pick another seed opposite that point in at least a few dimensions. The idea is to ensure that the seeds are well spread in at least a few dimensions. 

The algorithm for this method is defined below, and k means is run 30 times with these seeds below that. 

```python
def intializeclusters(ncluster, data):
    n= data.shape[0]
    dims=data.shape[1]
    dmids=np.zeros(dims)
    centers=np.zeros((ncluster,dims))
    
    for dindex in range(0,dims):
        col=data[:,dindex]
        dmax=max(col)
        dmin=min(col)
        dmids[dindex]=((dmax-dmin)/2)+dmin
    
    
    for i in range(0,ncluster):
        if i <= (ncluster+1)/2:
            centeri=np.random.randint(n)
            centers[i]=data[centeri]
            j=0
        variation=centers[j]-dmids

        maxdim=np.argmax(variation)
        mindim=np.argmin(variation)
        
        centeri=np.random.randint(n)
        centers[i]=data[centeri]
        centers[i][maxdim]=dmids[maxdim]-variation[maxdim]
        centers[i][mindim]=dmids[mindim]-variation[mindim]
        j+=1
    return centers
```

```python
iternum=30
i=0
total=len(kmeans.labels_)
ones=np.zeros(iternum)
zeros=np.zeros(iternum)
error=np.zeros(iternum)

for i in range(0,iternum):
        seeds=intializeclusters(2, np.array(data))
        kmeans = KMeans(n_clusters=2, init=seeds).fit(data)
        ones[i]=kmeans.labels_.sum()
        zeros[i]=total-ones[i]
        error[i]=meanSqDistance(np.array(data), kmeans.labels_, kmeans.cluster_centers_)
print(ones)
print(zeros)
print(error)
```
